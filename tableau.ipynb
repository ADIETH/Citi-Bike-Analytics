```python
# How has the proportion of short-term customers and annual subscribers changed?
df_customers = citi_bike_df.groupby('usertype')['usertype'].count().sort_values(ascending=False).reset_index(name='count')
print(df_customers.head())
x= np.array(df_customers[["count"]].values)
print("x=",x)
print("\nThe proportion of Annual Subscribers vs. short-term customers: \n", x*100.0/sum(x))
```

```python
citi_bike_df = citi_bike_df.dropna(how="any")
citi_bike_df.count() 
```

```python
# By what percentage has total ridership grown?   (doing it month on month)

print("Min starttime:",min(citi_bike_df["starttime"]))
print("Max starttime:",max(citi_bike_df["starttime"])) 

citi_bike_df["starttime"] = pd.to_datetime(citi_bike_df["starttime"])
citi_bike_df["Month_starttime"] = citi_bike_df["starttime"].map((lambda x: x.month))

citi_bike_df.Month_starttime.unique()

citi_bike_df.head() 
```

```python
# By what percentage has total ridership grown?   (doing it month on month)
print (citi_bike_df.groupby(["usertype","Month_starttime"])["tripduration"].count())
```

```python
df_per = citi_bike_df.groupby('Month_starttime')['Month_starttime'].count().reset_index(name='count')
df_per.sort_values(by='Month_starttime', ascending=False)

prev = df_per['count'][0]
for x in df_per['count'].tolist():
    print("prev=",prev)
    print("x = ",x)
    perc = (100/prev*x )-100
    print("percentage =", perc)
    prev = x
    #y = y + 1
                
df_per
```
